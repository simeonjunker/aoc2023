```python
with open('input') as f:
    data = f.read().strip().splitlines()
```

# Part 1

```python
X = 1
queue = [('start', d) for d in data]
i = 0

signal_strengths = []

while len(queue) > 0:
    if i+1 in [20, 60, 100, 140, 180, 220]:
        strength = X * (i+1)
        signal_strengths.append(strength)
    action, command = queue[0]
    queue = queue[1:]
    if action == 'start': 
        if command == 'noop':
            pass
        else: 
            queue = [('finish', command)] + queue
    else:
        _, V = command.split()
        X += int(V)
    i += 1

sum(signal_strengths)
```
